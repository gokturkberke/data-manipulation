```python
import seaborn as sns
tips=sns.load_dataset("tips")
df=tips.copy()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.lmplot(x="total_bill",y="tip",data=df);
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",data=df);
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="time",data=df);
```

```python
sns.lmplot(x="total_bill",y="tip",hue="smoker",col="time",row="sex",data=df);
```

ISI Haritasi

```python
import seaborn as sns
flights=sns.load_dataset('flights')
df=flights.copy()
df.head()
```

```python
df.shape
```

```python
df["passengers"].describe()
```

```python
df=df.pivot(index="month",columns="year",values="passengers");
df
```

```python
sns.heatmap(df);
```

```python
sns.heatmap(df, annot=True,fmt="d"); #annot true hucreler icinde sayilarin gozukmesi fmt=d ise tam sayi yani 5.0 degil 5 gozukmesini saglar
```

```python
sns.heatmap(df, annot=True,fmt="d",linewidths=.5);
```

Cizgi Grafik

```python
import seaborn as sns
fmri=sns.load_dataset("fmri")
df=fmri.copy()
df.head()
```

```python
df.shape
```

```python
df["timepoint"].describe()
```

```python
df["signal"].describe()
```

```python
df.groupby("timepoint")["signal"].count()
```

```python
#her bir zaman da esit miktarda signal toplanmis
```

```python
df.groupby("signal")["timepoint"].count()
```

```python
df.groupby("timepoint")["signal"].describe()
```

Cizgi Grafigin Olusturulmasi

```python
sns.lineplot(x="timepoint",y="signal",data=df);
```

```python
#mavi cizgiler ortalama , acik mavi de standart sapmalari olusturuyo(guven araligi)
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",data=df);
```

```python
sns.lineplot(x="timepoint",y="signal",hue="event",style="event",data=df);
```

```python
sns.lineplot(x="timepoint"
             ,y="signal"
             ,hue="event",
             style="event",
             markers=True,dashes=False,data=df); #dashes cizgilerin kesikli markers noktali gozukmesini saglar
```

```python
sns.lineplot(x="timepoint"
             ,y="signal"
             ,hue="region",
             style="event",
             data=df);
```
